### Webcraping thailand stock (SET) historical price
---
This notebook goal is to query **the historical price of all listed companies on Thailand stock market** into csv file.
<br>surce: https://finance.yahoo.com/
<br>list of tickers data from excel: 'list_of_set.xlsx'

```python
import time
import datetime
import pandas as pd
import os
```

#### Create new folder name <font color=blue>'price'</font> in current directory for store .csv file

```python
newpath = r'prices' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
else :
    print('this folder name already exist')
    pass
```

#### Function for download csv from yahoo.finance

```python
def set_hitorical_price_csv(name):
    ticker = name.upper() #Name Set Stock ex. Airasia type 'AAV'
    period1 = int(time.mktime(datetime.datetime(1970, 1, 1, 23, 59).timetuple())) # (YYYY, MM, DD, HH, mm)
    period2 = int(time.mktime(datetime.datetime(2021, 12, 29, 23, 59).timetuple()))
    present = int(time.time())
    interval = '1d' # 1d, 1m

    # query for period1 to period2
    #query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}.BK?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
    
    # query for period1 to present
    query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}.BK?period1={period1}&period2={present}&interval={interval}&events=history&includeAdjustedClose=true'
    df = pd.read_csv(query_string)
    # save dateframe into .csv file
    df.to_csv(f'prices/{ticker}_historical_price.csv',index=False)
```

#### Import dataframe form excel file

```python
sf = pd.read_excel('listedCompanies_th_TH.xlsx',skiprows=1)
sf
```

---
#### Create  batch of .csv file

```python
for i in sf.หลักทรัพย์:
    try :
        set_hitorical_price_csv(i)
        print(f'{i} Done')
    except:
        print(f'{i} Fail')
        pass
print('====End of process====')
```
